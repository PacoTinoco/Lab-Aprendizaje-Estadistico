```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

## Data batería :

```python
datos_bateria = pd.read_csv("C:\\Users\\Francisco\\Downloads\\8vo Semestre\\Lab estadistico\\bateria_tableta.txt")
x_bat = datos_bateria['carga'].values
y_bat = datos_bateria['duracion'].values
datos_bateria.head()
```

#### Regresiones :

```python
# Grado 1
X1 = x_bat.reshape(-1,1)
modelo1 = LinearRegression().fit(X1, y_bat)
r2_1 = r2_score(y_bat, modelo1.predict(X1))

# Grado 3
X3 = np.column_stack((x_bat, x_bat**2, x_bat**3))
modelo3 = LinearRegression().fit(X3, y_bat)
r2_3 = r2_score(y_bat, modelo3.predict(X3))

# Grado 5
X5 = np.column_stack((x_bat, x_bat**2, x_bat**3, x_bat**4, x_bat**5))
modelo5 = LinearRegression().fit(X5, y_bat)
r2_5 = r2_score(y_bat, modelo5.predict(X5))

# Grado 7
X7 = np.column_stack((x_bat, x_bat**2, x_bat**3, x_bat**4, x_bat**5, x_bat**6, x_bat**7))
modelo7 = LinearRegression().fit(X7, y_bat)
r2_7 = r2_score(y_bat, modelo7.predict(X7))

print(f'R² Grado 1: {r2_1:.4f}')
print(f'R² Grado 3: {r2_3:.4f}')
print(f'R² Grado 5: {r2_5:.4f}')
print(f'R² Grado 7: {r2_7:.4f}')
```

#### Graficar :

```python
x_plot = np.linspace(x_bat.min(), x_bat.max(), 200)

y1 = modelo1.predict(x_plot.reshape(-1,1))
y3 = modelo3.predict(np.column_stack((x_plot, x_plot**2, x_plot**3)))
y5 = modelo5.predict(np.column_stack((x_plot, x_plot**2, x_plot**3, x_plot**4, x_plot**5)))
y7 = modelo7.predict(np.column_stack((x_plot, x_plot**2, x_plot**3, x_plot**4, x_plot**5, x_plot**6, x_plot**7)))

plt.figure(figsize=(10,6))
plt.scatter(x_bat, y_bat, alpha=0.6, label='Datos originales')
plt.plot(x_plot, y1, label=f'Grado 1 (R²={r2_1:.4f})')
plt.plot(x_plot, y3, label=f'Grado 3 (R²={r2_3:.4f})')
plt.plot(x_plot, y5, label=f'Grado 5 (R²={r2_5:.4f})')
plt.plot(x_plot, y7, label=f'Grado 7 (R²={r2_7:.4f})')
plt.xlabel('Carga')
plt.ylabel('Duración')
plt.title('Batería Tableta - Regresiones Polinomiales')
plt.legend()
plt.grid(True)
plt.show()
```
