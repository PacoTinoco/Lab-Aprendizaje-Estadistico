```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = pd.read_csv('/content/Salarios_minimos.csv')
df.head()
```

```python
x_sal = df['Periodo']/1000
y_sal = df['Salario'].values
x_sal2= x_sal.values
```

```python
y_sal
```

```python
X1 = x_sal2.reshape(-1,1)
modelo1 = LinearRegression().fit(X1, y_sal)
r2_1 = r2_score(y_sal, modelo1.predict(X1))
```

```python
## escalar todas las variables MENOS LA Y para que esten en el dominio de 0-1 sin usar min max scaler:
X = boston.drop('MEDV', axis=1)
X.head()
```

```python
X_scaled = (X - X.min()) / (X.max() - X.min())
X_scaled.head()
```

```python
## distribucion z :
X_scaled_z = (X - X.mean()) / X.std()
X_scaled_z.describe()
```

```python
## ahora histogramas  :
X_scaled_z.hist(figsize=(15, 10))
```
