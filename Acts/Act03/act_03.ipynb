```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

df = pd.read_csv("C:\\Users\\Francisco\\Downloads\\8vo Semestre\\Lab estadistico\\Lab-Aprendizaje-Estadistico\\Data\\datos_feo_bien.csv")
```

```python
df.head()
```

### Hacer one hot con pandas :

```python
df1 = pd.DataFrame()
df1['tamaño'] = (df['tamaño'] == 'grande').astype(int)
df1['hielo'] = (df['hielo'] == 'con').astype(int)
df1['kahlua'] = (df['sabor'] == 'Kahlúa').astype(int)
df1['horchata'] = (df['sabor'] == 'horchata').astype(int)
df1['score'] = df['score']
df1
```

```python
#Regresion parte 1
X1 = df1[['tamaño', 'hielo', 'kahlua', 'horchata']]
X1 = sm.add_constant(X1)
y = df1['score']

modelo1 = sm.OLS(y, X1).fit()
print(modelo1.summary())
```

### Parte 2 :

```python
#Crear nuevo dataframe con codificacion -1/1
df2 = pd.DataFrame()

df2['tamaño'] = df['tamaño'].apply(lambda x: 1 if x == 'grande' else -1)
df2['hielo'] = df['hielo'].apply(lambda x: 1 if x == 'con' else -1)
df2['kahlua'] = (df['sabor'] == 'Kahlúa').astype(int)
df2['horchata'] = (df['sabor'] == 'horchata').astype(int)
df2['score'] = df['score']
```

```python
#Regresion parte 2
X2 = df2[['tamaño', 'hielo', 'kahlua', 'horchata']]
X2 = sm.add_constant(X2)
y2 = df2['score']

modelo2 = sm.OLS(y2, X2).fit()
print(modelo2.summary())
```

### Parte 3 comparar :

```python
print("Parte 1 (0/1):")
print(modelo1.params)
print()
print("Parte 2 (-1/1):")
print(modelo2.params)
print()
print("Parte 1 R2:", modelo1.rsquared)
print("Parte 2 R2:", modelo2.rsquared)
```

## OBSERVACIONES:

*  1. El R2 es igual en ambos casos, la capacidad predictiva es la misma

*  2. Los coeficientes de tamaño y hielo cambian entre las dos codificaciones:

   - En 0/1: el coeficiente representa el cambio de chica a grande (o sin a con)

    - En -1/1: el coeficiente representa la mitad del cambio total
#
*  3. El intercepto cambia porque la referencia es diferente:

   - En 0/1: el intercepto es el valor base cuando todo es 0

    - En -1/1: el intercepto es el valor promedio
#
*  4. Los coeficientes de kahlua y horchata no cambian porque usan la misma codificacion
#

* 5. La codificacion -1/1 es util cuando queremos interpretar el efecto como desviacion del promedio
