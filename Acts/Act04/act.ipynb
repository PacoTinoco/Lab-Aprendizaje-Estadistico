```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
import warnings
warnings.filterwarnings('ignore')
```

## Métodos de búsqueda :

+ Francisco Tinoco   |  Clase 23 Marzo 


Función dado el rango de [2, 15]

valores cuando x es 10, y = 2,847
x es 2, y = -4017
x es 3, y = -1255

```python
x = np.array([[10], [2], [3], [7], [12]])
y = np.array([[2847], [-4017], [-1255], [1773], [563]])
```

```python
x = np.linspace(2, 15, 10001).reshape(-1, 1)
```

```python
#kernel = kernels.RBF(length_scale=1.0)
gpr = GaussianProcessRegressor(n_restarts_optimizer= 10, random_state=42)
```

```python
gpr.fit(x,y)
```

```python
xsim = np.linspace(2, 15, 10001).reshape(-1, 1)
ysim, sigma = gpr.predict(xsim, return_std= True)
```

```python
plt.plot(xsim, ysim)
```

```python
## ahora crear y_upper = +- 2sigma

y_upper = ysim +2*sigma
```

```python
y_upper
```

```python
plt.plot(xsim, y_upper)
```

```python
xmax =np.argmax(y_upper)
xsim[xmax]
```
