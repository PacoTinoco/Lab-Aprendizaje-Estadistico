2. calcular   R2 de (1)

3. Usar 1 árbol de decision para regresion sobre el mismo dataset

4. Calcular R2 de (3)

* Usar la libreria de sklearn --> Decision Tree Regressor

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

df = pd.read_csv('../Data/Advertising.csv')
df.head()
```

```python
df = df.drop(columns=[col for col in df.columns if 'Unnamed' in col], errors='ignore')
```

### Sacamos variales para nuestros modelitos

```python
X = df[['TV', 'radio', 'newspaper']]
y = df['sales']
```

```python
# Regresión Lineal
lr = LinearRegression()
lr.fit(X, y)
y_pred_lr_all = lr.predict(X)
r2_lr_all = r2_score(y, y_pred_lr_all)
```

```python
# Árbol de Decisión
tree = DecisionTreeRegressor(random_state=42)
tree.fit(X, y)
y_pred_tree_all = tree.predict(X)
r2_tree_all = r2_score(y, y_pred_tree_all)
```

#### Con train y test :

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5, random_state=42
)

#RL
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
r2_lr = r2_score(y_test, y_pred_lr)

#Árbol
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)
r2_tree = r2_score(y_test, y_pred_tree)
```

```python
print("USANDO TODO EL DATASET ")
print("R2 Regresión Lineal:", r2_lr_all)
print("R2 Árbol de Decisión:", r2_tree_all)
```

```python
print(" TRAIN / TEST 50/50 ")
print("R2 Regresión Lineal:", r2_lr)
print("R2 Árbol de Decisión:", r2_tree)
```
