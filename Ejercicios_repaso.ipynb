+ 1.- Dos equipos juegan una serie de juegos (4 de 7, el que gane 4 juegos gana). Cada equipo tiene un 50% de probabilidad de ganar cada juego (no empates). ¿Cuál es la probabilidad de que la serie se vaya a 7 juegos? (5/16)


```python
import random
import numpy as np
rng = np.random.default_rng()
juegos_7 = 0

def serie():
  wins_a = 0
  wins_b = 0
  games = 0
  while (wins_a < 4) and (wins_b < 4) :
    winner = rng.integers(low=1, high=3)
    games += 1
    if winner == 1:
      wins_a += 1
    else:
      wins_b += 1

  if games >= 7:
    return 1
  else:
    return 0

for i in range(1000000):
  juegos_7 += serie()

proporcion = juegos_7/1000000
print(proporcion)
```


2.- Una en mil personas tienen una enfermedad H, y la prueba de la enfermedad sale correcta 98% de las veces cuando la persona está enferma. Por otro lado, si la persona está enferma la prueba tiene un error del 1%. Si alguien se hace una prueba y sale positiva, ¿cuál es la probabilidad de que tengan la enfermedad? (98/1097)

```python
rng = np.random.default_rng(seed=42)
```

```python
res = 0
positivos = 0
enfermos_positivos = 0


def prueba():
  enfermo = False
  prueba_positiva = False
  if rng.integers(1, 1001) == 1:
    enfermo = True
  ## SI ESTA ENFERMO
  if enfermo:
    if rng.integers(1, 101) in range(1, 99):
      prueba_positiva = True
  ## NO ESTA ENFERMO
  else:
    if rng.integers(1, 101) == 1:
      prueba_positiva = True
    else:
      prueba_positiva = False
  return prueba_positiva, enfermo
for i in range(1000000):
  prueba_positiva, enfermo = prueba()
  if prueba_positiva:
    positivos += 1
    if enfermo:
      enfermos_positivos += 1

print(enfermos_positivos/positivos)
```

 * 3.- Tres amigos diferentes te dijeron que está lloviendo, y cada persona tiene 1/3 de probabilidad de mentir. Si el pronóstico del clima dijo que la probabilidad de lluvia era 25%, ¿cuál es la probabilidad de que sí esté lloviendo? (8/11)

```python
np.random.seed(42)
n_simulaciones = 100000

# Contadores
veces_todos_dicen_lluvia = 0
veces_lluvia_y_dicen_lluvia = 0

for _ in range(n_simulaciones):
    # ¿Realmente está lloviendo? (25% probabilidad)
    lluvia_real = np.random.random() < 0.25


    amigos_dicen_lluvia = 0
    for _ in range(3):
        miente = np.random.random() < 1/3

        if lluvia_real and not miente:  # Llueve y dice verdad
            amigos_dicen_lluvia += 1
        elif not lluvia_real and miente:  # No llueve y miente
            amigos_dicen_lluvia += 1

    # ¿Los 3 dijeron que llueve?
    if amigos_dicen_lluvia == 3:
        veces_todos_dicen_lluvia += 1

        # ¿Y realmente estaba lloviendo?
        if lluvia_real:
            veces_lluvia_y_dicen_lluvia += 1

# Probabilidad: P(lluvia | 3 dicen lluvia)
probabilidad = veces_lluvia_y_dicen_lluvia / veces_todos_dicen_lluvia
probabilidad
```
