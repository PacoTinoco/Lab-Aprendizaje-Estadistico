```python
import pandas as pd

ruta = "C:\\Users\\Francisco\\Downloads\\8vo Semestre\\Lab estadistico\\Lab-Aprendizaje-Estadistico\\Data\\Advertising.csv"
df= pd.read_csv(ruta)
df.head()
```

```python
from sklearn.linear_model import LinearRegression
df_tv = df[['TV','sales']]
df_tv
```

```python
x_tv = df_tv['TV'].values.reshape(-1,1)
y_tv = df_tv['sales'].values

modelo = LinearRegression()
modelo.fit(x_tv, y_tv)
```

```python
print("Beta 0 (intercepto):", modelo.intercept_)
print("Beta 1 (pendiente):", modelo.coef_[0])
```
