## Ejercicio 1
- Utiliza el dataset Iris, modela con SVC y haz Cross-Validation de diferentes kernels (linear, poly, rbf, sigmoid).
- Modela con LogisticRegression.
- El metodo de Cross-Validation es K-Folds con k=10.
- Utiliza el AUC como metrico de Cross-Validation.
- Compara resultados.

```python
# Importar librerias
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = pd.read_csv("C:\\Users\\Francisco\\Downloads\\8vo Semestre\\Lab estadistico\\Lab-Aprendizaje-Estadistico\\Data\\IRIS.csv")
iris.shape
```

```python
iris.info()
```

```python
X_iris = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y_iris = iris['species']

#Codificar target a numeros
le = LabelEncoder()
y_iris_encoded = le.fit_transform(y_iris)

#Escalar features
scaler = StandardScaler()
X_iris_scaled = scaler.fit_transform(X_iris)

print("Clases:", le.classes_)
```

```python
#Cross-Validation con SVC para diferentes kernels
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
resultados_iris = {}
```

```python
print("RESULTADOS IRIS - SVC con diferentes kernels (K=10, AUC)")
print("="*60)

for kernel in kernels:
    modelo_svc = SVC(kernel=kernel, probability=True, random_state=42)
    scores = cross_val_score(modelo_svc, X_iris_scaled, y_iris_encoded, 
                             cv=10, scoring='roc_auc_ovr')
    resultados_iris[f'SVC_{kernel}'] = scores.mean()
    print(f"SVC ({kernel}): AUC = {scores.mean():.4f} (+/- {scores.std():.4f})")
```

```python
#Logistic Regression
modelo_lr = LogisticRegression(max_iter=1000, random_state=42)
scores_lr = cross_val_score(modelo_lr, X_iris_scaled, y_iris_encoded, 
                            cv=10, scoring='roc_auc_ovr')

resultados_iris['LogisticRegression'] = scores_lr.mean()
print(f"Logistic Regression: AUC = {scores_lr.mean():.4f} (+/- {scores_lr.std():.4f})")
```

```python
#Comparacion de resultados IRIS
print("COMPARACION DE MODELOS - IRIS")
print("="*60)

for modelo, auc in sorted(resultados_iris.items(), key=lambda x: x[1], reverse=True):
    print(f"{modelo:<25} AUC = {auc:.4f}")

mejor = max(resultados_iris, key=resultados_iris.get)
print(f"\nMejor modelo: {mejor} con AUC = {resultados_iris[mejor]:.4f}")
```

```python
#Preparar datos Default
default['student_num'] = (default['student'] == 'Yes').astype(int)

X_default = default[['balance', 'income', 'student_num']]
y_default = (default['default'] == 'Yes').astype(int)
```
